```python
#seleniumとbeautifulsoupをインポート
import chromedriver_binary
from selenium import webdriver
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager #追記
import time
from bs4 import BeautifulSoup
import requests
import bs4
import pandas as pd
```

```python
# レビューを取得。200ページ2000件で約50分（スリープ時間により変動）

pages = 200
title_list = []
body_list = []
review_point_list = []

#Google Chromeのドライバを用意
driver = webdriver.Chrome(ChromeDriverManager().install()) #追記

for page in range(pages):
    if page == 0:
        print(page,"AAA")
        url = 'https://www.tripadvisor.jp/Attraction_Review-g14129730-d320047-Reviews-Tokyo_Tower-Shibakoen_Minato_Tokyo_Tokyo_Prefecture_Kanto.html'
    else:
        print((page)*10, "BBB")
        url = f'https://www.tripadvisor.jp/Attraction_Review-g14129730-d320047-Reviews-or{(page)*10}-Tokyo_Tower-Shibakoen_Minato_Tokyo_Tokyo_Prefecture_Kanto.html'
# https://www.tripadvisor.jp/Attraction_Review-g14129730-d320047-Reviews-or10-Tokyo_Tower-Shibakoen_Minato_Tokyo_Tokyo_Prefecture_Kanto.html        
# https://www.tripadvisor.jp/Attraction_Review-g14129730-d320047-Reviews-or20-Tokyo_Tower-Shibakoen_Minato_Tokyo_Tokyo_Prefecture_Kanto.html
    driver.get(url)
    time.sleep(10)

    review_text = driver.find_elements_by_class_name("yCeTE") 
    time.sleep(1)
    review_point= driver.find_elements_by_class_name('UctUV')

    
    for x in range(0,len(review_text),2):
        title_list.append(review_text[x].text)
    for i in range(1,len(review_text),2):
        body_list.append(review_text[i].text)
    for i in range(45,55): #最新レビューが加わるごとに、はじめと終わりの数字をどちらも１つ増やす。現在（12/17）時点の最新レビューは、「楽しい！」 →→ ここも自動化できるといいね
        s = review_point[i].get_attribute('aria-label')[-3:].replace("'", "")
        # s = review_point[i].get_attribute('aria-label')[-10:].replace("'", "")
        review_point_list.append(s)
    print(review_point_list)
    print('-----')
    
```
